```python
import re
import requests
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
```

```python
# Read file list text file, where file links are stored
with open("../../rawdata/subset_M2TMNXAER_5.12.4_20240103_053124_.txt",'r') as f:
    urls = f.read()
```

```python
urls = urls.split('\n')
# The first PDF file readme is deleted
urls = urls[1:]
```

```python
# Using regular expression to create file names
# "M2TMNXAER.5.12.4" the data product I choose, you should open the file list text file and determine the name of 
# your chosen product
pattern = re.compile('M2TMNXAER.5.12.4/\d{4}/(.+).nc4')
files = [pattern.findall(url)[0] for url in urls]
```

```python
# As the NASA server can sometimes crash, we retry 3 times for each file
session = requests.Session()
retry = Retry(connect=3, backoff_factor=0.5)
adapter = HTTPAdapter(max_retries=retry)
session.mount('http://', adapter)
session.mount('https://', adapter)
```

```python
def down_load(url,file):
    """
    Function: Download files in the directory of the script
    Return: None
    """
    result = session.get(url)
    try:
        result.raise_for_status()
        f = open(file,'wb')
        f.write(result.content)
        f.close()
        print('contents of URL written to '+file)
    except:
        print('requests.get() returned an error code '+str(result.status_code))
```

```python
for url, file in zip(urls,files):
    down_load(url, file)
```
